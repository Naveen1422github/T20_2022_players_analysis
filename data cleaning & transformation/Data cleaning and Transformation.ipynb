```python
df_batting = pd.read_csv(r'C:\Users\praja\Downloads\shooting-range-master\Projects\T20_2022_players_analysis\data cleaning & transformation/batting_summary.csv')
```

```python
# removed extra index column
df_batting.drop(df_batting.columns[0], axis = 1, inplace = True)
df_batting
```

```python
# df_batting.rename(columns = {'out/not out':'dismissal'}, inplace = True)
```

```python
# df_batting['out/ not out'] = df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not out")
```

```python
# replacing extra text with out in out/not out column
df_batting['out/not out'] = df_batting['out/not out'].apply(lambda x: "out" if x != "not out" else x )
```

```python
# stroing scorecard/match_id in a dictionary so that i can connect each 
# table with this uniqu match id
match_id_dict = {}

for index, row in df_match.iterrows():
    key1 = row['Team 1'] + ' Vs ' + row['Team 2']
    key2 = row['Team 2'] + ' Vs ' + row['Team 1']
    match_id_dict[key1] = row['Scorecard']
    match_id_dict[key2] = row['Scorecard']
    
```

```python
# creating new column named match_id and filling values according to match with keys of match_id_dict
df_batting['match_id'] = df_batting['match'].map(match_id_dict)
```

**change scorecard to match**

```python
# renaming the scoredcard to match_id as it is confusing
df_match.rename(columns = {'Scorecard':'match_id'}, inplace = True) 
```
